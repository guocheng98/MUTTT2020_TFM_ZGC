```python
import transformers
import sentencepiece
```

```python
original = 'Helsinki-NLP/opus-tatoeba-es-zh'
fine_tuned = 'guocheng98/HelsinkiNLP-FineTuned-Legal-es-zh'
```

```python
tokenizer = transformers.MarianTokenizer.from_pretrained(original,use_fast=False)
model_original = transformers.MarianMTModel.from_pretrained(original)
model_FT = transformers.MarianMTModel.from_pretrained(fine_tuned)
```

```python
pipeline_original = transformers.TranslationPipeline(model_original,tokenizer,task='translation_es_to_zh')
pipeline_FT = transformers.TranslationPipeline(model_FT,tokenizer,task='translation_es_to_zh')
```

```python
f = open('Test_es.txt','r')
lines = f.readlines()
```

```python
original = open('Translation_original.txt','w')
FT = open('Translation_FT.txt','w')
for line in lines:
  translation_original = pipeline_original(line)
  translation_FT = pipeline_FT(line)
  text1 = translation_original[0]['translation_text']
  text2 = translation_FT[0]['translation_text']
  original.write(text1 + '\n')
  FT.write(text2 + '\n')
f.close()
original.close()
FT.close()
```
